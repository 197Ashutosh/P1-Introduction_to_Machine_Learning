```python
mean_petal_length = df_iris['petal length (cm)'].mean()
std_petal_length = df_iris['petal length (cm)'].std()

print("Mean value of petal length (cm):", mean_petal_length) 
print("Standard deviation of petal length (cm):", std_petal_length)
print("Minimum value of petal length (cm):", df_iris['petal length (cm)'].min())
print("Maximum value of petal length (cm):", df_iris['petal length (cm)'].max())
```

```python
upper_limit = mean_petal_length + 3 * std_petal_length
lower_limit = mean_petal_length - 3 * std_petal_length

print("Upper limit:", upper_limit)
print("Lower limit:", lower_limit)
```

```python
df_outliers = df_iris[(df_iris['petal length (cm)'] > upper_limit) | (df_iris['petal length (cm)'] < lower_limit)]
print("Outliers:")
print(df_outliers)
```

```python
df_iris_cleaned = df_iris[(df_iris['petal length (cm)'] <= upper_limit) & (df_iris['petal length (cm)'] >= lower_limit)]


print(df_iris_cleaned.describe())


sns.histplot(df_iris_cleaned, kde=True, multiple="stack")
plt.show()
```

```python
Q1 = df_iris['petal length (cm)'].quantile(0.25)
Q3 = df_iris['petal length (cm)'].quantile(0.75)
```

```python
IQR = Q3 - Q1

print("Q1 (25th percentile) of petal length (cm):", Q1)
print("Q3 (75th percentile) of petal length (cm):", Q3)
print("Interquartile Range (IQR) of petal length (cm):", IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
```
